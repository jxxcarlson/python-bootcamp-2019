```python
def read_file(file_name):
    file = open(file_name)
    data = file.read()
    file.close()
    return data

def string_to_lines(data):
    return data.split("\n")

def print_list(list):
    i = 0
    for item in list:
        print(i, item)
        i = i + 1
```

```python
data_as_string = read_file('sea_level.txt')
len(data_as_string)
```

```python
data_as_string[:600]
```

```python
data_as_lines = string_to_lines(data_as_string)
len(data_as_lines)
```

```python
print_list(data_as_lines[:10])
```

```python
print_list(data_as_lines[52:60])
```

```python
data_good_lines = data_as_lines[52:]
```

```python
data_good_lines[0], data_good_lines[-1]
```

```python
line = data_good_lines[0]
line.split(" ")
```

```python
def parse_line(str):
    return list(filter(lambda x: x != '', str.split(" ")))
    
parse_line(line)    
# http://book.pythontips.com/en/latest/map_filter.html
```

```python
parsed_lines = list(map(parse_line, data_good_lines))
parsed_lines[0]
```

```python
parsed_lines2 = list(filter(lambda x: (len(x) > 0 and x[0] != '999'), parsed_lines))
parsed_lines2[:10]
```

```python
def line_to_data(line):
    return [float(line[2]), float(line[11])]

line_to_data(parsed_lines2[0])
```

```python
data = list(map(line_to_data, parsed_lines2))
```

```python
data
```

```python
from matplotlib import pyplot as plt
```

```python
xs = list(map(lambda x: x[0], data))
ys = list(map(lambda x: x[1], data))
```

```python
plt.plot(xs, ys)
```

```python
import numpy as np
m, b = np.polyfit(xs, ys, 1)
def y(x):
    return m*x + b
```

```python
fitted_line_ys = list(map(y, xs))
```

```python
plt.plot(xs, ys)
plt.plot(xs, fitted_line_ys, color = 'red')
plt.show()
plt.savefig('sealevel.png')
```

```python
list(map(y, [2020, 2030,2040, 2050, 2060]))
```
